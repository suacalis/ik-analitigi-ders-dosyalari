```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv("http://suayipc.net/Veri/ikveri/df.csv", sep=',')
#df=pd.read_excel('http://suayipc.net/Veri/ikveri/df.xlsx')
```

```python
df.head()
```

```python
df= df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
df=pd.DataFrame(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pointplot(x="Yaş", y="Puan", data=df);
```

```python
plt.figure(figsize=(10,5))
plt.title('Yaşa Göre Puanlar')
sns.pointplot(x="Yaş", y="Puan", data=df);
```

```python
plt.figure(figsize=(10,6))
plt.title('Yaşa Göre Puanlar')
sns.barplot(x='Yaş', y='Puan', color='black', data=df); 
```

```python
plt.figure(figsize=(20,7))
plt.title('Yaşa Göre Puanlar')
sns.barplot(x='Yaş', y='Puan', hue='Meslek',data=df); 
```

```python
plt.figure(figsize=(10,6))
plt.title('Meslek')
sns.countplot(x="Meslek", data=df);
```

```python
plt.figure(figsize=(12,7))
plt.title('İsim ')
sns.countplot(x="İsim", data=df);
```

```python
sns.lmplot(x="Yaş", y="Puan", data=df);
```

```python
sns.boxplot(data = df);
```
