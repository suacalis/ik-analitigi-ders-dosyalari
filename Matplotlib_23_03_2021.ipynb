```python
plt.figure(figsize=(9,4))
plt.bar(df.Yaş, df.Puan)
plt.title("Yaşa Göre Puan")
plt.xlabel("Yaş")
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.bar(df['Yaş'], df['Puan'])
plt.title("Yaşa Göre Puan")
plt.xlabel("Yaş")
plt.ylabel("Puan")
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.barh(df['Yaş'], df['Puan'])
plt.title("Yaşa Göre Puan")
plt.xlabel("Yaş")
plt.ylabel("Puan")
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.scatter(df.Yaş, df.Puan)
plt.title("Yaşa Göre Puan")
plt.xlabel("Yaş")
plt.ylabel("Puan")
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.scatter(df.ÇocukSayısı, df.Puan)
plt.title("Çocuk Sayısına Göre Puan")
plt.xlabel("Çocuk Sayısı")
plt.ylabel("Puan")
plt.show()
```
